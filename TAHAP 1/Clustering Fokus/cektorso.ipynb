Cek Torso (Centroind)

AFI

```python
import os, glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- 1. List semua file torso untuk Afi ---
pattern_afi = os.path.join(FOLDER_TORSO, "Afi_*_torso.csv")
torso_files_afi = sorted(glob.glob(pattern_afi))

print("Daftar file torso Afi:")
for i, f in enumerate(torso_files_afi):
    print(f"{i}: {os.path.basename(f)}")

# --- 2. PILIH salah satu file Afi yang mau diplot ---
# ganti index di sini kalau mau file lain
idx_afi = -7
afi_torso_csv = torso_files_afi[idx_afi]
print("\nFile yang dipakai:", os.path.basename(afi_torso_csv))

df_afi = pd.read_csv(afi_torso_csv)

# pastikan terurut
df_afi = df_afi.sort_values('frame').reset_index(drop=True)

# --- 3. Plot torso_x vs frame ---
plt.figure(figsize=(8, 3.5))
plt.plot(df_afi['frame'], df_afi['torso_x'], marker='o', linewidth=1)
plt.xlabel("Frame")
plt.ylabel("Torso X (m)")
plt.title(f"Trajectory torso_x vs frame | {os.path.basename(afi_torso_csv)}")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 4. Plot torso_y vs frame ---
plt.figure(figsize=(8, 3.5))
plt.plot(df_afi['frame'], df_afi['torso_y'], marker='o', linewidth=1)
plt.xlabel("Frame")
plt.ylabel("Torso Y (m)")
plt.title(f"Trajectory torso_y vs frame | {os.path.basename(afi_torso_csv)}")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 5. Plot torso_z vs frame ---
plt.figure(figsize=(8, 3.5))
plt.plot(df_afi['frame'], df_afi['torso_z'], marker='o', linewidth=1)
plt.xlabel("Frame")
plt.ylabel("Torso Z (m)")
plt.title(f"Trajectory torso_z vs frame | {os.path.basename(afi_torso_csv)}")
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 6. Hitung v_walk_rel (Y per frame) ---
df_afi['dy'] = df_afi['torso_y'].diff()
df_afi['v_walk_rel'] = df_afi['dy']   # Y per frame (belum dikali fps)

plt.figure(figsize=(8, 3.5))
plt.plot(df_afi['frame'], df_afi['v_walk_rel'], marker='o', linewidth=1)
plt.axhline(0, color='k', linewidth=0.8)
plt.xlabel("Frame")
plt.ylabel("v_walk_rel (Y per frame)")
plt.title("Kecepatan relatif berjalan (approx v_walk) per frame - Afi")
plt.grid(True)
plt.tight_layout()
plt.show()
```
